```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
print("What Age group people lived more?")
print("Most passengers were aged 20-40")
```

```python
sns.countplot(x="2urvived", data=df)
plt.title("Survival Count")
plt.xlabel("Survived(1=Yes, 0=No)")
plt.ylabel("Count")
plt.show()
print("Does more no. of passengers survived")
print("Yes, More passengers did not survive than those who did.")
```

```python
sns.countplot(x="2urvived", hue="Sex", data=df)
plt.title("Survival Count by Gender")
plt.xlabel("Survived (1=Yes, 0=No)")
plt.ylabel("Count")
plt.legend(title="Gender", labels=["Female", "Male"])
plt.show()
print("who has the most survival rate")
print("Insight: Females had a higher survival rate than males.")
```

```python
sns.barplot(x="Pclass", y="2urvived", data=df)
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Passenger Class")
plt.show()
print("Check survival rates for each passenger class (1st, 2nd, 3rd class).")
print("Insights: Higher-class passengers had a better survival rate.")
```

```python
df["age_group"] = pd.cut(df["Age"], bins=[0, 12, 18, 60, 100], labels=["Child", "Teen", "Adult", "Elderly"])
sns.barplot(x="age_group", y="2urvived", data=df)
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Age Group")
plt.show()
print("Categorize passengers into children, adults, and elderly.")
print("Insights: Children had a higher survival rate due to evacuation priority.")
```

```python
sns.boxplot(x="2urvived", y="Fare", data=df)
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare Price")
plt.title("Fare Price vs Survival")
plt.show()
print("Check if higher fare passengers had a better survival rate.")
print("Insights: Higher fare passengers had a better chance of survival.")
```

```python
df["family_size"]=df["sibsp"]+df["Parch"]
sns.barplot(x="family_size",y="2urvived", data=df)
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Family Size")
plt.show()
print("Did people travelling alobe survive less?")
print("Passengers with small families had better survival rates")
```
